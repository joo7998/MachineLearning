```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1.신경망에서 많이 쑈이는 sigmoid 함수

```python
def sigmoid_function(x):
    return 1/(1+np.exp(-x))

x=np.linspace(-5,5)
y=sigmoid_function(x)

plt.plot(x,y)
plt.show()
```

sigmoid 함수의 출력값은 항상 : 0<y<1 

## 2.tanh
- it solves just one problem of being zero-centred.
y = (e<sup>x</sup> - e<sup>-x</sup>) / (e<sup>x</sup> + e<sup>-x</sup>)

```python
def tanh_function(x):
    return np.tanh(x)

x=np.linspace(-5,5)
y=tanh_function(x)

plt.plot(x,y)
plt.show()
```

tanh 함수 : -1<y<1

## 3.ReLU(Rectified Linear Unit) 함수
- widely used with Convolutional Neural networks
- (+) easy to compute & not saturate and does not cause the Vanishing Gradient Problem
- (-) not being zero centred, “dying ReLU” problem(the output is zero for all negative inputs), sometimes leads to unusable nodes
- 
- f(x) = max(0,x)
- y = if x <= 0 then else x  (0 이상이면 입력값 그대로, 0 이하면 0을 출력=> dying ReLU)

```python
def relu_function(x):
    return np.where(x <= 0, 0, x)

x=np.linspace(-5,5)
y=relu_function(x)

plt.plot(x,y)
plt.show()
```

## 4.Leaky ReLU
- x가 음수인 영역에서 아주 작은 기울기 생성 
- 출력이 0이 되어, 더 이상 학습이 진행되지 않는 뉴런이 발생하는 dying ReLU 현상을 피하기 위해 사용 (if we set α as 1 then Leaky ReLU will become a linear function f(x) = x and will be of no use)
- 
- f(x) = max(αx, x)    ====> α is a hyperparameter generally set to 0.01
- y = if x <= 0 then 0.01x else x

# 여기서부터 https://towardsdatascience.com/everything-you-need-to-know-about-activation-functions-in-deep-learning-models-84ba9f82c253 다시 읽기

```python
def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x)

x=np.linspace(-5,5)
y=leaky_relu_function(x)

plt.plot(x,y)
plt.show()
```

## 5.SoftMax
- 입력받은 값을 출력으로 0~1 사이의 값으로 모두 정규화 
- 출력 값들의 총합은 항상 1
- a more generalised form of the sigmoid. 
- used in multi-class classification problems. 
- Similar to sigmoid, it produces values in the range of 0–1 => used as the FINAL LAYER in classification models.

- k번일 확률/전체확률
- n=총 클래스의 수, k=k번째 클래스

```python
def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

y=softmax_function(np.array([1,2,3]))
print(y)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = softmax_function(x)

fig = plt.figure(figsize=(10,7))
fig.set_facecolor('white')

plt.plot(x,y)
plt.ylim(0,0.1)
plt.show()
```

## 7.활성화 함수로 Sigmoid 함수 활용
- Only for historical reasons and never used in real models. 
- Computationally expensive, causes vanishing gradient problem and not zero-centred. 
- Generally used for binary classification problems.

```python
x1 = np.arange(-1.0, 1.0, 0.2) # -1~1 사이 0.2 단위로
x2 = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))

w_x1 = 2.5
w_x2 = 3.0

bias = 0.1

for i in range(10):
    for j in range(10):
        u = x1[i]*w_x1 + x2[j]*w_x2 + bias
        y = 1/(1 + np.exp(-u))
        Z[j][i] = y

plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

```python
# numpy 내 행렬 곱셈 이용 (the simpler, the better)

x1 = np.arange(-1.0, 1.0, 0.2) # -1~1 사이 0.2 단위로
x2 = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))

W = np.array([2.5, 3.0])

bias = np.array([0.1])

for i in range(10):
    for j in range(10):
        list1 = [x1[i], x2[j]]
        x = np.array(list1)
        u = np.dot(x, W.T) + bias
        y = 1/(1 + np.exp(-u))
        Z[j][i] = y

plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

##### Multi-Layer Perceptron (3 Layers)
- 입력층 (Input Layer):
- 은닉층 (Hidden Layer): row = n개 
- 출력층 (Output Layer): column : 3개 ()
- perceptron은 1차 함수 기반 

# 8.Machine Learning의 Deep Learning : Definition  

- 1개 이상의 은닉층을 사용하는, 신경망 네트위크 모델

- 입력층 (Input Layer): 2
- 은닉층 (Hidden Layer): node 2개, Activation Functions Sigmoid
- 출력층 (Output Layer): node 2개, Activation Functions 항등함수 

```python
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)
Z = np.zeros((10,10))

w_im = np.array([[4.0, 4.0], [4.0, 4.0]]) # 출력은 열의 갯수 만큼나옴 = 2개
w_mo = np.array([[1.0], [-1.0]]) # 출력은 열의 갯수 만큼나옴 = 1개

b_im = np.array([3.0, -3.0])
b_mo = np.array([0.1])

def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    return 1 / (1 + np.exp(-u))

def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return u

for i in range(10):
    for j in range(10):
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
        Z[j][i] = out[0]
        
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```
